# Single Leading Underscore: _var

```python
# The underscore prefix is meant as a hint to 
# another programmer that a variable or method 
# starting with a single underscore is intended 
# for internal use. 
```

```python
# when you import module 
# it adds all functions (simple and underscores) 
import my_module
my_module.internal()
my_module._external()
```

```python
# when you module => from import * 
# it doesn't add (underscored functions)
from my_module import *
internal()
```
